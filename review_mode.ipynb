读论文 ；

第一层：快速浏览（15分钟）

读标题、摘要和结论 - 了解论文解决了什么问题和主要结果
浏览图表和表格 - 图表往往直观呈现核心结果
阅读小标题和章节首句 - 了解论文结构


第二层：理解核心（30-45分钟）

深入阅读引言和相关工作 - 了解问题背景和意义
仔细阅读方法部分 - 聚焦于理解算法核心思想，而非每个细节
分析主要结果 - 确认结果是否支持论文主张


第三层：深度分析（按需进行）

研究技术细节和推导 - 按照自己的节奏理解复杂部分
查看参考文献 - 对关键概念有疑问时深入挖掘
增强理解的实用技巧

主动阅读法

提出问题 - 阅读前列出几个你希望论文回答的问题
做笔记和标注 - 在论文上标记关键点，写下自己的想法
用自己的话重述 - 每读完一个部分，用自己的语言总结

论文标题：

核心问题：论文试图解决什么问题？

关键创新：论文的主要贡献是什么？

方法概述：算法/方法如何工作？（简明描述或流程图）

主要结果：最重要的发现是什么？

个人思考：这对我的工作有何启发？有什么局限性？

跟进问题：需要进一步了解的概念或方法
